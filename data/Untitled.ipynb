# NUMPY 
* numpy.full(A,B) = 균일값 생성(A개만큼의 B)
* numpy.zeros(A,dtype = int) = A개만큼의 0 생성
* numpy.ones(A,dytpe = int) = A개만큼의 1 생성
* numpy.random.random(A) = A개만큼의 랜덤값 생성
* numpy.arange(A,B,C) = A부터 B-1까지의 C간격으로 범위값 생성
* array[[A,B]] = A번 인덱스 , B번 인덱스를 리스트로 받아옴
* array1[[2:11:2]] = 2번인덱스부터 10번 인덱스까지 간격 2로 받아옴
* array.std() = 표준 편차
* array.var() = 분산

# pandas
## column = column , row = index
* .dtypes = column의 자료형 반환
### 인덱싱
* .loc[a,b] = low = a column = b => 값 받아옴 
    - a만 있으면 그 줄에 있는걸 다받아옴
    - b는 A.loc[] 형식이 아닌 A[b] 형식으로 해야 출력됨
#### column 추가 
    -딕셔너리마냥 추가하면됨 a[b] = c
####  삭제
    - iphone_df.drop(제목 , axis = 'index'or 'column', inplace => 기존 데이터프레임 변경여부 )
#### column index 이름바꾸기
    - rename
#### 대표 지정
    - set_index
## .describe()
    -컬럼 정보 다나옴
### .head(a) , .tail(a), .shape, .columns, .info
    - 앞에서 몇개, 뒤에서 몇개 , (row,column) 출력 ,각 컬럼정보
## .sort_values(by= A , ascending = B, inplace = C)
    - A를 기준으로 , B = True 면 오름차순, inplace True면 원래배열에 적용
### .unique()
    -겹치는거 없앰
### .value_counts()
    - unique + 카운트
## .corr()
    -상관계수
